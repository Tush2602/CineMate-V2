```python
import requests
import matplotlib.pyplot as plt
from PIL import Image
from io import BytesIO
import os

# 🔑 Replace with your TMDB API key
TMDB_API_KEY = os.getenv("TMDB_API_KEY", "")

def show_movie_poster(movie_id):
    # Step 1: Get movie details
    url = f"https://api.themoviedb.org/3/movie/{movie_id}?api_key={API_KEY}"
    response = requests.get(url)
    
    if response.status_code != 200:
        print("Error fetching movie data")
        return
    
    data = response.json()
    
    # Step 2: Get poster path
    poster_path = data.get("poster_path")
    
    if not poster_path:
        print("No poster available")
        return
    
    # Step 3: Build full image URL
    base_url = "https://image.tmdb.org/t/p/w500"
    full_poster_url = base_url + poster_path
    
    # Step 4: Fetch image
    img_response = requests.get(full_poster_url)
    img = Image.open(BytesIO(img_response.content))
    
    # Step 5: Show using matplotlib
    plt.imshow(img)
    plt.axis('off')
    plt.title(data.get("title", "Movie Poster"))
    plt.show()


# 🎬 Example usage
show_movie_poster(284052)  # Fight Club
```
